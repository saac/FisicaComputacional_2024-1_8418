```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# %matplotlib notebook
# from ipywidgets import *

from pylab import *
import math as mt
```

```python
def f(a,b,x,y):
    
    x1 = x
    y1 = y
    
    x = y1*(a+x1**2)
    y = b/(a+x1**2)
    
    return x,y
```

```python
a = 1
b = 2

x0 = 1.
y0 = 1.

x,y = x0,y0
for n in range(70):
    x,y = f(a,b,x,y)
    print("n =",n,"  x = ",x,"  y = ",y)

```

$x = y(a+x^2)$

$y = \dfrac{b}{(a+x^2)}$

$y = \dfrac{x}{(a+x^2)}$

$x = \sqrt{\dfrac{b}{y}-a}$

```python
def f2(a,b,x,y):
    
    x1 = x
    y1 = y
    
    x = np.sqrt(b/y1 - a)
    y = x1/(a+x1**2)
    
    return x,y
```

```python
a = 1
b = 2

x0 = 1.
y0 = 1.

x,y = x0,y0
for n in range(100):
    x,y = f2(a,b,x,y)
    print("n =",n,"  x = ",x,"  y = ",y)

```
